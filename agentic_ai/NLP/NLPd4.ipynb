pipeline1-sentiment analysis, classifies text as positive or negative with a confidence score.

```python
from transformers import pipeline
senti = pipeline('sentiment-analysis')
reviews = [
'This movie was absolutely breathtaking! A masterpiece.',
'Terrible acting, boring plot. Wasted 2 hours of my life.',
'It was okay. Nothing special but not bad either.',
'Shah Rukh Khan was brilliant. Loved every minute!',
'Ram is more attractive than jayanth ani evaro chepparu',
'KSRM is a waste college in asia',
'charan is bad '
]
results = senti(reviews)
for review, result in zip(reviews, results):
  label = result['label']
  score = result['score']
  print(f'{label} ({score:.1%}) | {review[:50]}...')
```

pipeline 2-text classification

```python
from transformers import pipeline
classifier= pipeline('zero-shot-classification',
model='facebook/bart-large-mnli')
text = 'ksrm is a college of engineering . in aiml we have best students in non crt but cse are worst except few(nen itey aswitha ni analey)'
labels = ['technology','course','computer science','artificial intellignece']
result = classifier(text, candidate_labels=labels)
for label, score in zip(result['labels'], result['scores']):
   print(f'{label:15} {score:.1%}')
```

```python
from transformers import pipeline

summariser = pipeline(
    task="summarization",
    model="facebook/bart-large-cnn"
)

long_text = """
My name is Suraiya. I am currently pursuing a B.Tech degree in Computer Science and Engineering.
I am interested in Artificial Intelligence, Machine Learning, Data Science, and Web Development.
I have completed certifications in Python and Human Computer Interaction.
I enjoy working on projects related to AI, Machine Learning, and Smart Agriculture solutions.
I am passionate about learning new technologies and improving my programming skills.
My goal is to become an AI Engineer and contribute to innovative technology solutions in the future.
"""

summary = summariser(
    long_text,
    max_length=20,
    min_length=10,
    do_sample=False,
    truncation=True
)

print(summary[0]["summary_text"])
```

**Pipeline 4 — Question Answering**
Extractive QA: given a context passage and a question, the model finds and returns the exact span of
text that answers the question.

```python
from transformers import pipeline

qa = pipeline(
    'question-answering',
    model='deepset/bert-base-cased-squad2'
)

context = '''
Suraiya is pursuing a B.Tech degree in Computer Science and Engineering.
She is interested in Artificial Intelligence, Machine Learning, and Data Science.
She has completed certifications in Python and Human Computer Interaction.
She enjoys working on AI-based projects and smart agriculture solutions.
Her goal is to become an AI Engineer.
'''

questions = [
    'What degree is Suraiya pursuing?',
    'What are Suraiya’s areas of interest?',
    'Which certifications has Suraiya completed?'
]

for q in questions:
    answer = qa(question=q, context=context)

    print(f'Q: {q}')
    print(f'A: {answer["answer"]} (confidence: {answer["score"]:.1%})')
```

**Pipeline 5 — Translation**

```python
from transformers import pipeline

translator = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M"
)

texts = [
    "My name is Suraiya.",
    "I am pursuing a B.Tech degree in Computer Science and Engineering.",
    "I am interested in Artificial Intelligence and Machine Learning.",
    "My goal is to become an AI Engineer."
]

for text in texts:
    result = translator(
        text,
        src_lang="eng_Latn",
        tgt_lang="tel_Telu"
    )

    print("EN:", text)
    print("TE:", result[0]["translation_text"])
    print()
```
