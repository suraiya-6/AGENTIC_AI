Pipeline 1 — Sentiment Analysis Classifies text as POSITIVE or NEGATIVE with a confidence score. Uses DistilBERT fine-tuned on the SST-2 movie reviews dataset by default.

```python
from transformers import pipeline

senti = pipeline("sentiment-analysis")

reviews = [
    "This movie was absolutely breathtaking! A masterpiece.",
    "Terrible acting, boring plot. Wasted 2 hours of my life.",
    "It was okay. Nothing special but not bad either.",
    "Shah Rukh Khan was brilliant. Loved every minute!",
    "Ram is more attractive than Jayanth ani evaro chepparu",
    "KSPM is a great college in Asia",
    "Charan is bad"
]

results = senti(reviews)

for review, result in zip(reviews, results):
    score = result["score"]

    # Custom labels
    if score >= 0.90:
        sentiment = "Strong"
    elif score >= 0.60:
        sentiment = "Moderate"
    else:
        sentiment = "Neutral"

    print(f"{sentiment} ({score:.1%}) | {review}")
```

Pipeline 2 — Text Classification (Zero-shot) Zero-shot classification lets you define your OWN category labels — the model was never trained on your specific categories but still classifies correctly by understanding the semantic meaning of your labels.

```python
from transformers import pipeline
# Zero-shot: classify without any task-specific training
classifier= pipeline('zero-shot-classification',
model='facebook/bart-large-mnli')
text = 'ksrm is a college of engineering . in aiml we have best students in non crt but cse are worst except few(nen itey aswitha ni analey)'
labels = ['technology','course','computer science','artificial intellignece']
result = classifier(text, candidate_labels=labels)
for label, score in zip(result['labels'], result['scores']):
   print(f'{label:15} {score:.1%}')
```

```python
from transformers import pipeline

summariser = pipeline(
    task="summarization",
    model="facebook/bart-large-cnn"
)

long_text = """
My name is Shaik Iqra. I am passionate about Machine Learning, Artificial Intelligence,
and Natural Language Processing. I have been learning Python and various machine
learning algorithms such as Linear Regression, Logistic Regression, KNN, Naive Bayes,
Decision Trees, Random Forest, and K-Means Clustering. Recently, I started exploring
Natural Language Processing concepts including tokenization, stopword removal,
stemming, lemmatization, Bag of Words, TF-IDF, and transformer models. I enjoy
building projects and solving real-world problems using data science techniques.
My goal is to become a skilled AI and Machine Learning Engineer and contribute to
innovative technologies in the future.
"""

summary = summariser(
    long_text,
    max_length=30,
    min_length=10,
    do_sample=False
)

print(summary[0]["summary_text"])
```

Pipeline 4 — Question Answering Extractive QA: given a context passage and a question, the model finds and returns the exact span of text that answers the question.




```python
from transformers import pipeline

qa = pipeline(
    'question-answering',
    model='deepset/bert-base-cased-squad2'
)

context = '''
KSRM College of Engineering (KSRMCE) is located in Kadapa, Andhra Pradesh.
It is one of the well-known engineering colleges in the region.
The college offers undergraduate and postgraduate programs in various engineering disciplines.
KSRMCE focuses on technical education, innovation, research, and student development.
The college provides modern laboratories, library facilities, and placement support for students.
'''

questions = [
    'Where is KSRM College of Engineering located?',
    'What programs does KSRMCE offer?',
    'What does KSRMCE focus on?',
    'What facilities are available at KSRMCE?'
]

for q in questions:
    answer = qa(question=q, context=context)

    print(f'Q: {q}')
    print(f'A: {answer["answer"]} (confidence: {answer["score"]:.1%})')
    print()
```

Pipeline 5 — Translation




```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Step 3: Load Translation Model and Tokenizer directly
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Step 4: English Sentences
texts = [
    "My name is Shaik Iqra.",
    "I am pursuing a B.Tech degree in Computer Science and Engineering.",
    "I am interested in Artificial Intelligence and Machine Learning.",
    "My goal is to become an AI Engineer."
]

# Step 5: Translate English → Telugu
for text in texts:
    # Encode the source text with the specified source language
    tokenizer.src_lang = "eng_Latn"
    encoded_input = tokenizer(text, return_tensors="pt")

    # Generate translation IDs with the target language
    generated_tokens = model.generate(
        **encoded_input,
        forced_bos_token_id=tokenizer.convert_tokens_to_ids("tel_Telu") # Corrected to use convert_tokens_to_ids
    )

    # Decode the generated tokens
    translated_text = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]

    print("EN:", text)
    print("TE:", translated_text)
    print()
```
