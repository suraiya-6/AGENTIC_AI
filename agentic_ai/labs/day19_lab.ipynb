 Lab: Build a study-buddy chatbot that answers questions about any topic you give it

```python
# Study Buddy Chatbot

print("================================")
print("     STUDY BUDDY CHATBOT")
print("Type 'exit' to quit")
print("================================")

knowledge_base = {
    "python": "Python is a high-level programming language used for web development, AI, data science, and automation.",

    "machine learning": "Machine Learning is a branch of AI that enables computers to learn from data and make predictions.",

    "data science": "Data Science combines statistics, programming, and domain knowledge to extract insights from data.",

    "nlp": "Natural Language Processing enables computers to understand and process human language.",

    "database": "A database is an organized collection of data stored electronically.",

    "ai": "Artificial Intelligence is the simulation of human intelligence in machines."
}

while True:
    question = input("\nAsk a topic: ").lower()

    if question == "exit":
        print("Thank you for using Study Buddy Chatbot!")
        break

    elif question in knowledge_base:
        print("\nAnswer:")
        print(knowledge_base[question])

    else:
        print("\nSorry! I don't know that topic.")
        print("Try: Python, AI, NLP, Database, Machine Learning, Data Science")
```
