#Lab: Student Report Card System Using CSV Files

```python
import pandas as pd

# Read CSV file
df = pd.read_csv("students.csv")

# Calculate Total and Percentage
df["Total"] = (
    df["Maths"] +
    df["Science"] +
    df["English"] +
    df["Computer"] +
    df["Social"]
)

df["Percentage"] = df["Total"] / 5

# Function to assign grades
def assign_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "F"

# Create Grade column
df["Grade"] = df["Percentage"].apply(assign_grade)

# Display Report Card
print("\n===== STUDENT REPORT CARD =====\n")
print(df[["Student_ID", "Name", "Total", "Percentage", "Grade"]])

# Save output to CSV
df.to_csv("report_card.csv", index=False)

print("\nReport card generated successfully!")
print("Output saved as report_card.csv")
```
